```python
from bertopic import BERTopic
from sentence_transformers import SentenceTransformer
import umap
import hdbscan
import pandas as pd
```

```python
interactions_list = gymitrainer['content'].tolist()[:6000]
```

```python
import spacy

nlp = spacy.load("xx_ent_wiki_sm")

# Choose which entity types to exclude
EXCLUDE_ENT_TYPES = {"PERSON", "ORG"}

from tqdm import tqdm

def remove_named_entities_from_list(texts, exclude={"PERSON", "ORG", "GPE", "LOC", "PRODUCT", "WORK_OF_ART"}):
    cleaned = []
    for doc in tqdm(nlp.pipe(texts, batch_size=4), total=len(texts), desc="Removing named entities"):
        cleaned.append(" ".join([token.text for token in doc if token.ent_type_ not in exclude]))
    return cleaned


interactions_list = remove_named_entities_from_list(interactions_list)
```

```python
len(interactions_list)
```

```python
st_model = SentenceTransformer('paraphrase-multilingual-mpnet-base-v2', device='mps', )
```

```python
embeddings = st_model.encode(interactions_list, show_progress_bar=True, batch_size=16,
                             convert_to_tensor=False)
print(f"Number of docs: {len(embeddings)}, vector size: {len(embeddings[0])}")

```

```python
np.save("./data/embeddings.npy", embeddings)
```

```python
embeddings = np.load("./data/embeddings.npy", allow_pickle=True)
```

```python
umap_model = umap.UMAP(metric='cosine', random_state=42, n_components=10)
# reduced_embeddings = umap_model.fit_transform(embeddings) shouldn't do this due to it being done in bertopic internally
```

```python
hdbscan_model = hdbscan.HDBSeCAN(
    min_cluster_size=15,
    metric='euclidean',
    cluster_selection_method='eom',
    prediction_data=True
)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from nltk.corpus import stopwords

stop_words = stopwords.words('german')  
vectorizer = CountVectorizer(stop_words=stop_words)
```

```python
topic_model = BERTopic(
    umap_model=umap_model,        # From step 3
    hdbscan_model=hdbscan_model,  # From step 4
    vectorizer_model=vectorizer,  # Optional; CountVectorizer for keyword extraction
    calculate_probabilities=True, # Enable soft clustering outputs
    verbose=True                  # Show progress in terminal
)
```

```python
topics, probs = topic_model.fit_transform(interactions_list, embeddings)
```

```python
topic_model.visualize_topics()
```

```python
topic_model.visualize_barchart()
```

```python
topic_model.visualize_heatmap()
```

```python
used_chatbot_ids = set(all_scores.user_id.unique()) & set(mapping[mapping['confidence']>0.5].user_id.unique())
```

```python
no_chatbot_ids = set(all_scores.user_id.unique()) - set(mapping[mapping['confidence']>0.5].user_id.unique())
```

```python
chatbotter_scores = all_scores[all_scores['user_id'].isin(used_chatbot_ids)]
keinchatbotter_scores = all_scores[~all_scores['user_id'].isin(used_chatbot_ids)]
```
